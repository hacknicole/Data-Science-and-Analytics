**A company started to invest in digital marketing as a new way of their product
promotions. For that they collected data and decided to carry out a study on it.**

● The company wishes to clarify whether there is any increase in sales after stepping into digital marketing.

● The company needs to check whether there is any dependency between
the features “Region” and “Manager”.

**Help the company to carry out their study with the help of data provided.**

#Increase in Sales after stepping into Digital Marketing

```python
f_stat, p_value = f_oneway(data_sales['Sales_before_digital_add(in $)'], data_sales['Sales_After_digital_add(in $)'])
```

```python
print('Value of p: ',p_value)
```

```python
if p_value < 0.05:
  print('Reject Null Hypothesis')
else:
  print('Failure to Reject Null Hypothesis')
```

Since the value of p is large, there is an increase in sales after stepping into digital marketing

# Dependency between the features "Region" and "Manager"

```python
fet_table = pd.crosstab(data_sales['Region'], data_sales['Manager'])
```

```python
fet_table
```

```python
chi2_stat, p_value, dof, exp = chi2_contingency(fet_table)
```

```python
print('Value of p: ',p_value)
```

```python
if p_value < 0.05:
  print('Reject Null Hypothesis')
else:
  print('Failure to Reject Null Hypothesis')
```

Since the value of p is small, the null hypothesis is rejected in failure as there is no dependency betweeen "Region" and "Manager".
