1. Write a python program to create an array and covert to a
desired dimension, as shown in picture.

Original array:

[[8  9  10]

[11  12  13]]

Reshaped array:

[[ 8  9]

[10  11]

```python
org_arr = np.array(([8,9,10],[11,12,13]))
print('Original array:')
print(org_arr)
print("\r")
re_arr = org_arr.reshape(3,2)
print('Reshaped array')
print(re_arr)
```

4. Read the given data set and answer the following
questions. Show the steps taken to produce your final
answer.

a. Display the Glucose level and the Age for the
samples having the index ranging from 45 to 60.

b. How many patients in the data are having the blood
pressure level above 120?

c. Find the minimum and maximum value of insulin
level in the data.

d. Create a histogram of "Age".

e. Create a scatter plot of "BMI" vs "Glucose

```python
dia_data = pd.read_csv('/content/diabetes_model.csv')
dia_data
```

```python
#a. Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.
dia_data.loc[45:60,['Glucose','Age']]
```

```python
#b. How many patients in the data are having the blood pressure level above 120?
bp = (dia_data['BloodPressure']>120).sum()
print('Number of patients having blood pressure above 120: ',bp)
```

```python
#c. Find the minimum and maximum value of insulin level in the data.
min = dia_data['Insulin'].min()
max = dia_data['Insulin'].max()

print('Minimum value of insulin level: ',min)
print('Maximum value of insulin level: ',max)
```

```python
#d. Create a histogram of "Age".
import matplotlib.pyplot as plt
age_plt = plt.hist(dia_data['Age'])
age_plt
```

```python
#e. Create a scatter plot of "BMI" vs "Glucose
bmi_vs_glc = plt.scatter(dia_data['BMI'],dia_data['Glucose'])
plt.title('BMI vs Glucose')
plt.xlabel('BMI')
plt.ylabel('Glucose')
bmi_vs_glc
```
