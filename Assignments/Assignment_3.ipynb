```python
import pandas as pd
```

1. Read the dataset to python environment.

```python
data = pd.read_excel('/content/iris.xls')
print('Iris Dataset')
data
```

2. Display the columns in the dataset.

```python
columns = data.columns
print(columns)
```

3. Calculate the mean of each column of the dataset.

```python
col_mean =data[['SL','SW','PL','PW']].mean()
print('Mean Value of each column of the dataset: ')
print(col_mean)
```

4. Check for the null values present in the dataset


```python
data_nullvalues = data.isnull().sum()
print('Null Values: ')
print(data_nullvalues)
```

5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
import matplotlib.pyplot as plt
```

```python
#Bar plot of SL vs Classification in Iris Data
plt.bar(data['SL'],data['Classification'])
plt.title('SL vs Classification')
plt.xlabel('SL')
plt.ylabel('Classification')
```

```python
# Histogram of Classification in Iris Data
plt.hist(data['Classification'])
```

```python
# Boxplot of SL, SW, PL and PW in Iris Data
plt.boxplot(data[['SL','SW','PL','PW']])
plt.show()
```

```python
# Seaborn Pairplot of Iris Data
import seaborn as sns
sns.pairplot(data)
```

```python
# Scatter plot of Iris Data
plt.scatter(data['SL'],data['SW'],data['PL'],data['PW'])
```
