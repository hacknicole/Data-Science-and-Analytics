4. What do you think about the scoring variation for math, reading and writing
based on

â— Gender

â— Test preparation course


##**GENDER**

```python
gender_var = data.groupby('gender')['math score', 'reading score', 'writing score'].std()
gender_var
```

```python
# Plotting the scoring variation based on Gender
gender_var.plot(kind = 'bar')
plt.title('Scoring Variation based on Gender')
plt.xlabel('Gender')
plt.xticks(rotation=0)
plt.ylabel('Scoring Variation')
plt.legend(title = 'Scores')
```

Here, the scoring variation in three tests for male students does not affect much whereas, in female students, the scoring variation for math test affects the most.

##**TEST PREPARATION COURSE**

```python
testprep_var = data.groupby('test preparation course')['math score','reading score','writing score'].std()
testprep_var
```

```python
# Plotting the scoring variation based on Test Preparation Course
testprep_var.plot(kind = 'bar')
plt.title('Scoring Variation based on Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.xticks(rotation = 0)
plt.ylabel('Scoring Variation')
plt.legend(title = 'Scores')
```

The scoring variation for both the completed tests and none tests does not have much variation.
