Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset into python environment and make ‘PassengerId’ as the index column

```python
data = pd.read_csv('titanic_dataset.csv', index_col='PassengerId')
data
```

Fill in all the missing values present in all the columns in the dataset

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
data.drop('Cabin',axis=1,inplace=True)
```

```python
data
```

```python
data.isna().sum()
```

Handling the outliers in at least 3 columns in the dataset using zscore

```python
from scipy.stats import stats,zscore
z_score = zscore(data[['Age','Fare','SibSp','Parch']])
z_score
```

```python
threshold =  3
data_new = data[(z_score < threshold).all(axis=1)]
print(data_new)
```

```python
data_new
```

Doing min max scaling on the feature set taking 'Survived' as the target


```python
X_data = data.drop(['Survived','Name','Sex','Ticket','Embarked'], axis=1)
X_data
```

```python
X_data.describe()
```

```python
X_data.hist(figsize=(20,15))
plt.show()
```

```python
import sklearn.preprocessing as sk
minmax = sk.MinMaxScaler()
X_mm = minmax.fit_transform(X_data)
X_mm = pd.DataFrame(X_mm)
X_mm.describe()
```

```python
X_mm.hist(figsize=(20,15))
plt.show()
```
